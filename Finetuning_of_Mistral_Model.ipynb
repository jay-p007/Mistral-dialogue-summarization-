```python
from transformers import AutoTokenizer, AutoModelForCausalLM, GPTQConfig, TrainingArguments, GenerationConfig
```

```python
processing_class = AutoTokenizer.from_pretrained("TheBloke/Mistral-7B-Instruct-v0.1-GPTQ")
```

```python
processing_class
```

```python
processing_class.eos_token
```

```python
processing_class.pad_token = processing_class.eos_token
```

```python
quantization_config_loading = GPTQConfig(bits=4, disable_exllama=True, tokenizer = processing_class)
```

```python
model = AutoModelForCausalLM.from_pretrained("TheBloke/Mistral-7B-Instruct-v0.1-GPTQ",
                                             quantization_config=quantization_config_loading,
                                             device_map="auto")
```

```python
print(model)
```

```python
model.config.use_cache = False
model.config_pretraining_tp = 1
model.gradient_checkpointing_enable()
```

```python
from peft import prepare_model_for_kbit_training
```

```python
model = prepare_model_for_kbit_training(model)
```

```python
from peft import LoraConfig
```

```python
peft_config = LoraConfig(
    lora_alpha=16,
    lora_dropout=0.05,
    r=16,
    bias="none",
    task_type="CAUSAL_LM",
    target_modules=["q_proj", "v_proj"]
)
```

```python
from peft import get_peft_model
```

```python
model = get_peft_model(model, peft_config)
```

```python
from trl import SFTTrainer
```

```python
from transformers import TrainingArguments
```

```python
training_arguments = TrainingArguments(
    output_dir="mistral-finetuned-samsum",
    per_device_train_batch_size=8,
    gradient_accumulation_steps=1,
    optim="paged_adamw_32bit",
    learning_rate=2e-4,
    lr_scheduler_type="cosine",
    save_strategy="epoch",
    num_train_epochs=1,
    logging_steps=100,
    max_steps=250,
    fp16=True,
    push_to_hub=True,
    report_to="none",
)
```

```python
# Define trainer
trainer = SFTTrainer(
    model=model,
    processing_class=processing_class,
    args=training_arguments,
    train_dataset=data,
    peft_config=peft_config,
)
```

```python
#Model Training
trainer.train()
```

```python
#Trained model push to Hub
trainer.push_to_hub()
```

# **Inferencing (General Model Vs Finetuned Model ):**

```python
from transformers import AutoModelForCausalLM
from transformers import AutoTokenizer, GenerationConfig
import torch
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

# Load General Mistral-7B Model
general_model_name = "TheBloke/Mistral-7B-Instruct-v0.1-GPTQ"
general_model = AutoModelForCausalLM.from_pretrained(general_model_name, torch_dtype="auto", device_map="auto",offload_folder="offload")
general_tokenizer = AutoTokenizer.from_pretrained(general_model_name)
```

```python
# Load Fine-Tuned Mistral Model
finetuned_model_name = "/content/mistral-finetuned-samsum"
finetuned_model = AutoModelForCausalLM.from_pretrained(finetuned_model_name, torch_dtype=torch.float16, device_map="auto", offload_folder="offload")
finetuned_tokenizer = AutoTokenizer.from_pretrained(finetuned_model_name)
```

```python
def generate_summary(model, tokenizer, dialogue):
    prompt = f"Summarize the following conversation:\n\n{dialogue}\n\nSummary:"
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
    output = model.generate(**inputs, max_new_tokens=100, pad_token_id=tokenizer.eos_token_id)
    return tokenizer.decode(output[0], skip_special_tokens=True)
```

```python
sample_dialogue = """
Alice: Hey, did you finish the report?
Bob: Not yet, but I’ll send it soon.
Alice: We have a meeting at 3 PM, can you send it before that?
Bob: Yes, I’ll wrap it up in 30 minutes.
Alice: Great, thanks!
"""
```

```python
# Inference using General Mistral-7B
general_summary = generate_summary(general_model, general_tokenizer, sample_dialogue)

# Inference using Fine-Tuned Mistral-7B
finetuned_summary = generate_summary(finetuned_model, finetuned_tokenizer, sample_dialogue)

# Print Results
print("🔹 General Mistral-7B Summary:\n", general_summary)
print("\n🔹 Fine-Tuned Mistral-7B Summary:\n", finetuned_summary)
```
